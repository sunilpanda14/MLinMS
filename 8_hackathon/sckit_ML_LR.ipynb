```python
from autogluon.tabular import TabularDataset, TabularPredictor
data = TabularDataset(df)
data
```

```python
data['Rupture_Results'] = data['Rupture_Results'].str.strip()
```

```python
# Split the data into train and test datasets
from sklearn.model_selection import train_test_split

D_train, D_test = train_test_split(data, test_size=0.20, random_state=42)
```

```python
predictor = TabularPredictor(label='Rupture_Results').fit(D_train)
```
